```python
import numpy as np
import pandas as pd
```

```python
#imported the companies data frame with encoding
companies = pd.read_csv("companies.txt", sep="\t", encoding = "ISO-8859-1")
```

```python
#details of companies data frame
print(companies.info())
print('\n')
print(companies.describe())
print('\n')
print(companies.shape)
print('\n')
print(companies.columns)
```

```python
# To remove special charcters from the columns 
companies['permalink'] = companies.permalink.str.encode('utf-8').str.decode('ascii', 'ignore')
companies['region'] = companies.region.str.encode('utf-8').str.decode('ascii', 'ignore')
```

```python
#converted the permalink to lower case
companies['permalink'] = pd.Series(map(lambda x : x.lower(),companies['permalink']))
```

```python
#companies unique values based on permalink
companies['permalink'].unique()
```

```python
# Count of unique values permalink in companies dataframe
len(companies['permalink'].unique())
```

```python
# Total records having unique values of company name . As we can see the unique value are less the total rows in the company 
# Name has duplicate values
len(companies['name'].unique())
```

```python
# Import then rounds2.csv got error so used encoding and inspeecting the data frame 
rounds2 = pd.read_csv("rounds2.csv", encoding = "ISO-8859-1")
print(rounds2.info())
print('\n')
print(rounds2.describe())
print('\n')
print(rounds2.shape)
print('\n')
print(rounds2.columns)
```

```python
# To remove special characters from the company_permalink column
rounds2['company_permalink'] = rounds2.company_permalink.str.encode('utf-8').str.decode('ascii', 'ignore')
rounds2
```

```python
# count of unique values 
# We see that with a flaw as seen there are some values in company_permalink in completely upper case while other in complete lower case
len(rounds2['company_permalink'].unique())
```

```python
# convert the company_permalink to completely lower case
rounds2['company_permalink'] = pd.Series(map(lambda x : x.lower(),rounds2['company_permalink']))
rounds2['company_permalink']
```

```python
# count the unique values for company_permalink after converting to lower case.The count is matching with the companies dataframe
len(rounds2['company_permalink'].unique())
```

```python
# To check whether there are any companies in the rounds2 file which are not  present in companies
rounds2[~rounds2.company_permalink.isin(companies.permalink)]
```

```python
# During merging got error as there was no common column in both dataframe. So added the below column in rounds2
rounds2['permalink'] = rounds2['company_permalink']
rounds2
```

```python
# used the above created column for merging of the two dataframe
master_frame = pd.merge(rounds2, companies, how='inner', on='permalink')
```

```python
# details of the newly created dataframe
print(master_frame.info())
print('\n')
print(master_frame.describe())
print('\n')
print(master_frame.shape)
print('\n')
print(master_frame.columns)
```

```python
# code for number of the rows 
len(master_frame.index)
```

```python
#column-wise null percentages here
print(round(100*(master_frame.isnull().sum()/len(master_frame.index)),2))
```

```python
# Dropping rows where the amount raised is null
master_frame = master_frame.drop(master_frame.loc[master_frame.raised_amount_usd.isnull() , :].index, axis = 0)
# to check if the row are dropped 
master_frame.loc[master_frame.raised_amount_usd.isnull(), :]
```

```python
# column-wise null percentages here
print(round(100*(master_frame.isnull().sum()/len(master_frame.index)),2))
```

```python
# Dropping the columns as they are not required for further analysis
master_frame = master_frame.drop(['founded_at','homepage_url'], axis=1)
```

```python
# To check the number of columns remaning
print(master_frame.info())
```

```python
# Describe the funding_round_code column which has maximum null values
master_frame['funding_round_code'].describe()
```

```python
# Finding the funding mode used maximum times having maximum funds raised
master_frame.groupby(['funding_round_code']).mean()
```

```python
# converting funding_round_code to type 'category'
master_frame['funding_round_code'] = master_frame['funding_round_code'].astype('category')

# displaying frequencies of each category in funding_round_code
master_frame['funding_round_code'].value_counts()
```

```python
# As per the above analysis we see funding_round_code as'A' is used maximum times.
# So Replacing funding_round_code null with 'A'
master_frame.loc[pd.isnull(master_frame['funding_round_code']),['funding_round_code']] = 'A'
```

```python
# To check no null values for funding_round_code
master_frame.loc[pd.isnull(master_frame['funding_round_code']),:]
```

```python
# column-wise null percentages here
print(round(100*(master_frame.isnull().sum()/len(master_frame.index)),2))
```

```python
# For further analysis we required values in country_code & category_list.
# We can see there are rows having null values for the above columns which is not desired.
#Drop the null values for country_code & category_list as they are used for further analysis 
master_frame = master_frame.drop(master_frame.loc[(master_frame.country_code.isnull()) | (master_frame.category_list.isnull()) , :].index, axis = 0)
# To check if the rows are deleted
master_frame.loc[(master_frame.country_code.isnull()) | (master_frame.category_list.isnull()) , :]
```

```python
# column-wise null percentages here
print(round(100*(master_frame.isnull().sum()/len(master_frame.index)),2))
# As we see below most of the other values null values are state_code,region,city which cannot be defaulted.
# Also we cannot remove these rows as they have country_code present and will be used for further analysis
# So we have not handled the null values for the remaning columns
```

```python
# Adding Primary Sector in the Master Dataframe based on the Category_list
master_frame['primary_sector'] = master_frame['category_list'].str.split('|',1).str[0]
# To check primary_sector added properly
master_frame.head(10)
```

```python
# Loading the mapping csv in the datframe
mapping = pd.read_csv("mapping.csv")
```

```python
# Checking details for mapping csv
mapping.info()
```

```python
# there are many names in the category list having na as 0 . So replacing 0 to na
mapping['category_list'] = mapping['category_list'].str.replace('0','na')
```

```python
# Creating a new dataframe using the melt function
mapping2=pd.melt(mapping,id_vars=['category_list'],var_name='main_sector', value_name='Val')
mapping2
```

```python
# Soring the mapping2 dataframe with the required details(Correct Main Sector and Primary Sector)
mapping2 = mapping2.loc[mapping2.Val == 1]
```

```python
# Check the mapping2 dataframe
mapping2
```

```python
# Creating a new dataframe with required columns
sector =pd.DataFrame(mapping2.loc[:,['category_list','main_sector']])
```

```python
sector = sector.rename(columns={'category_list':'primary_sector'})
```

```python
# Merge the two datframes to get the main_sector in the master_frame
master_frame = pd.merge(master_frame, sector, how='inner', on='primary_sector')
```

```python
master_frame.info()
```

```python
# Creating Country Specific Dataframes
#USA
D1 = pd.DataFrame(master_frame.loc[(master_frame.funding_round_type == 'venture')  & (master_frame.country_code == 'USA') & (master_frame.raised_amount_usd>=5) & (master_frame.raised_amount_usd<=15) ,:])
#GBR
D2 = pd.DataFrame(master_frame.loc[(master_frame.funding_round_type == 'venture')  & (master_frame.country_code == 'GBR') & (master_frame.raised_amount_usd>=5) & (master_frame.raised_amount_usd<=15) ,:])
#IND
D3 = pd.DataFrame(master_frame.loc[(master_frame.funding_round_type == 'venture')  & (master_frame.country_code == 'IND') & (master_frame.raised_amount_usd>=5) & (master_frame.raised_amount_usd<=15) ,:])
```

```python
# Code for counting the the number investments in each sector  and  country   
print("USA")
print(pd.DataFrame({'count' : D1.groupby( [ "main_sector"] ).size()}).reset_index().sort_values(by='count',ascending=False))
print("\n")
print ("Total count of investment in USA ", len(D1.index))
print("\n")
print("GBR")
print(pd.DataFrame({'count' : D2.groupby( [ "main_sector"] ).size()}).reset_index().sort_values(by='count',ascending=False))
print("\n")
print ("Total count of investment in GBR ", len(D2.index))
print("\n")
print("IND")
print(pd.DataFrame({'count' : D3.groupby( [ "main_sector"] ).size()}).reset_index().sort_values(by='count',ascending=False))
print ("Total count of investment in IND ", len(D3.index))
```

```python
# Print total amount invested in each country 
# Country 1 : USA
print("USA")
print(D1.reset_index().groupby( [ "main_sector"],as_index=False ).sum().sort_values(by='raised_amount_usd', ascending=False))
print("\n")
print ("Total amount of investment in USA ", D1.raised_amount_usd.sum())
print("\n")

# Country 2 : GBR
print("GBR")
print(D2.reset_index().groupby( [ "main_sector"],as_index=False ).sum().sort_values(by='raised_amount_usd', ascending=False))
print("\n")
print ("Total amount of investment in GBR ", D2.raised_amount_usd.sum())
print("\n")
# Country 3 : IND
print("IND")
print(D3.reset_index().groupby( [ "main_sector"],as_index=False ).sum().sort_values(by='raised_amount_usd', ascending=False))
print ("Total amount of investment in IND ", D3.raised_amount_usd.sum())
print("\n")
```
